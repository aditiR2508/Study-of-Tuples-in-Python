```python
## Python Tuples-->type of an array
#Creating a Tuple and Accessing its Items
Theatre = ("act", "costumes", "escape", "entertainment", "stage", "props", "set")
print(Theatre[4])
```

```python
#Slicing in Tuples
Theatre = ("act", "costumes", "escape", "entertainment", "stage", "props", "set")
print(Theatre[2:5])
print(Theatre[1:6:2])
print(Theatre[-2:-6:-1])
print(Theatre[1::5])
```

```python
#Tuples are similar to Lists except for the fact that they cannot be made changes to, however slicing is possible
## Adding items
Theatre = ("act", "costumes", "escape", "entertainment", "stage", "props", "set")
Theatre[2] = "skits"
print(Theatre)
## This function returns an error because tuples do not support making changes to them
```

```python
#appending tuples
Theatre = ("act", "costumes", "escape", "entertainment", "stage", "props", "set")
Theatre.append("skits")
print(Theatre)
## This will also return an error
```

```python
#Problem 1: A student's exam result is stored in a tuple containing three fixed fields: subject name, marks obtained and grade.
## Python program to: Store exam result using tuple; Unpack tuple elements into separate variables; Display subject, marks and grade;
## Check whether student has scored 75 or more marks and print "Distinction" if condition is satisfied.

ExamRes= ("physics", 89, "A") #storing exam result

subject, marks, grade = ExamRes #unpacking into separate variables

print("Subject Name: ", subject) #displaying fields
print("Marks: ", marks)
print("Grade: ", grade)

if marks >=75:  #checking if condition
 print("Distinction")
```
